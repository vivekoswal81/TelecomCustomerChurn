```python
def conv_sen(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv_sen)
```

```python
tot_counts=df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100

#plot

fig, ax = plt.subplots(figsize=(6,6)) # adjust figsize for better visualization

# plot the bars

tot_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4','#ff7f0e']) 

#add percentage labels on the 
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y =p.get_xy()
    ax.text(x+width/2, y+height/2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='center')
plt.show()
```

#greater percentage of senior citizen customers have churned out
